```python
"""
Multi-provider LLM switcher for LangChain practice.

Setup:
    pip install langchain langchain-openai langchain-google-genai python-dotenv

.env file should contain:
    GITHUB_PAT=ghp_xxx
    GEMINI_API_KEY=xxx
    GROQ_API_KEY=xxx   (optional, for later)
"""

from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from langchain_google_genai import ChatGoogleGenerativeAI
import os

load_dotenv()

# ---------------------------------------------------------
# 1. Change this one variable to switch providers
# ---------------------------------------------------------
PROVIDER = "groq"   # options: "github", "gemini", "groq"


def get_llm(provider: str):
    if provider == "github":
        # GitHub Models — free, rate-limited, good general prototyping
        return ChatOpenAI(
            base_url="https://models.github.ai/inference",
            api_key=os.environ["GITHUB_PAT"],
            model="openai/gpt-4o",
        )

    elif provider == "gemini":
        # Google Gemini — free tier, big context window
        return ChatGoogleGenerativeAI(
            google_api_key=os.environ["GEMINI_API_KEY"],
            model="gemini-2.5-flash",
        )

    elif provider == "groq":
        # Groq — free tier, very fast inference, OpenAI-compatible
        return ChatOpenAI(
            base_url="https://api.groq.com/openai/v1",
            api_key=os.environ["GROQ_API_KEY"],
            model="llama-3.3-70b-versatile",
        )

    else:
        raise ValueError(f"Unknown provider: {provider}")


if __name__ == "__main__":
    llm = get_llm(PROVIDER)

    prompt = "Explain what an AI agent is in one sentence."
    response = llm.invoke(prompt)

    print(f"--- Provider: {PROVIDER} ---")
    print(response.content)
```
